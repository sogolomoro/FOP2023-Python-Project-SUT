```python
def to_pascal_case(name):
    return ''.join(word.capitalize() for word in name.split('_'))
pascal_case_names = {column: to_pascal_case(column) for column in column_names}
```

نمودار Pie Chart (نمودار دایره‌ای یا نمودار پای) یک نمودار مدلسازی‌کننده است که از یک دایره برای نمایش دادن تقسیم بندی یک مجموعه داده به بخش‌ها استفاده می‌کند. هر بخش از دایره نمایانگر یک دسته یا دسته از داده‌ها است و نسبت مساحت بخش به مساحت کل دایره، نسبت فراوانی یا درصد آن دسته از داده‌ها را نشان می‌دهد.

کاربردهای نمودار Pie Chart عبارتند از:

نمایش توزیع درصدی: نمودار Pie Chart به خوبی برای نمایش توزیع درصدی اجزاء یک مجموعه مناسب است. به عنوان مثال، درصد فروش محصولات یک شرکت یا توزیع تعداد اتاق‌های یک هتل.

نمایش تقسیم بندی: این نمودار برای نشان دادن تقسیم بندی یک مجموعه به دسته‌های مختلف استفاده می‌شود. به عنوان مثال، تقسیم بندی افراد در یک گروه بر اساس مهارت‌های مختلف.

نمایش اطلاعات جزئی: زمانی که تعداد اجزاء نسبتاً کم و به دسته‌های مختلف یا دسته‌بندهای اصلی قابل تفکیک است، از نمودار Pie Chart می‌توان برای نمایش این اطلاعات استفاده کرد.
